**IMPORT ALL NECESSARY MODULES**

```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from  sklearn import model_selection
from sklearn.metrics import classification_report
```

**READ IN DATAFRAME**

```python
df = pd.read_csv(f'C:/Users/user/Desktop/Churn_Modelling.csv')
df.head()
```

**EXPLORE DATA AND WRANGLE THE DATA**

```python
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
df
```

```python
df.isnull().sum(), df.dtypes
```

```python
print(df['Gender'].unique())
print(df['Gender'].value_counts())
```

```python
df1 = pd.get_dummies(df, columns=['Gender','Geography'], drop_first=True)
df1
```

```python
cols_to_scale = ['Tenure','EstimatedSalary','Balance', 'CreditScore']
scaler = MinMaxScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
df1
```

```python
X = df1.drop(['Exited'], axis='columns')
y = df1.Exited
y.value_counts()
```

**SPLIT INTO TRAINING AND TESTING SET**

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
x_train.shape, y_train.value_counts()
```

**MODEL BUILDING**

```python
def ANN (x_train,x_test,y_train,y_test):
    model = keras.Sequential([
        keras.layers.Dense(100, input_shape=(x_train.shape[1],), activation='relu'),
        keras.layers.Dense(50, activation='relu'),
        keras.layers.Dense(25, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
        ])

    # opt = keras.optimizers.Adam(learning_rate=0.01)

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    model.fit(x_train, y_train, epochs=50)
    y_pred = model.predict(x_test)
    y_pred = np.round(y_pred)
    print('Classification report:   \n', classification_report(y_test, y_pred))
```

```python
ANN(x_train,x_test,y_train,y_test)
```

**HANDLING IMBALANCE IN DATASET TO IMPROVE F1 SCORE**

1. UNDERSAMPLING THE MAJORITY CLASS

```python
df_not_exited = df1[df1.Exited == 0]
df_exited = df1[df1.Exited == 1]

count_exited = df_exited.shape[0]
count_not_exited = df_not_exited.shape[0]
count_exited, count_not_exited
```

```python
df_not_exited_undersample = df_not_exited.sample(count_exited)
df_not_exited_undersample.shape

```

```python
undersampled_combined_df = pd.concat([df_not_exited_undersample, df_exited], axis=0)
undersampled_combined_df['Exited'].value_counts()
```

```python
X = undersampled_combined_df.drop(['Exited'], axis=1)
y = undersampled_combined_df.Exited

x_train_US,x_test_US,y_train_US,y_test_US = model_selection.train_test_split(X, y, test_size=0.2, random_state=5, stratify=y)
```

```python
ANN(x_train_US,x_test_US,y_train_US,y_test_US)
```

2. OVERSAMPLING OF MINORITY CLASS

```python
count_exited, count_not_exited
```

```python
df_exited_oversample = df_exited.sample(count_not_exited, replace=True)
oversampled_combined_df = pd.concat([df_exited_oversample, df_not_exited], axis=0)

X = oversampled_combined_df.drop(['Exited'], axis=1)
y = oversampled_combined_df.Exited

x_train_OS,x_test_OS,y_train_OS,y_test_OS = model_selection.train_test_split(X, y, test_size=0.2, random_state=5, stratify=y)
```

```python
ANN(x_train_OS,x_test_OS,y_train_OS,y_test_OS)
```

3. USING SMOTE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
X = df1.drop(['Exited'], axis='columns')
y = df1.Exited
y.value_counts()
```

```python
X_sm, y_sm = smote.fit_resample(X, y)
y_sm.value_counts()
```

```python
x_train_sm,x_test_sm,y_train_sm,y_test_sm = model_selection.train_test_split(X_sm, y_sm, test_size=0.2, random_state=5, stratify=y_sm)
ANN(x_train_sm,x_test_sm,y_train_sm,y_test_sm)
```

FROM COMPARING ALL THE ABOVE, SMOTE HAS BETTER PERFORMANCE.
